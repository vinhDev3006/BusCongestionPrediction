```python
import numpy as np
import pandas as pd
import tensorflow as tf
import os
import matplotlib.pyplot as plt
```

```python
FILE_PATH = os.path.join("dataset", "data_with_segment_congestion_level.csv")

df = pd.read_csv(FILE_PATH)
```

```python
df.info()
```

```python
df.sort_values('window')
attr = df.columns[2]
label = attr
plt.hist(df[df['segment_congestion_level'] == 0][label], color='green', label="Free", alpha=0.7)
plt.hist(df[df['segment_congestion_level'] == 1][label], color='cyan', label="Basically Free", alpha=0.7)
plt.hist(df[df['segment_congestion_level'] == 2][label], color='blue', label="Mild Congestion", alpha=0.7)
plt.hist(df[df['segment_congestion_level'] == 3][label], color='orange', label="Moderate Congestion", alpha=0.7)
plt.hist(df[df['segment_congestion_level'] == 4][label], color='red', label="Heavy Congestion", alpha=0.7)
plt.title(label)
plt.ylabel("No.")
plt.xlabel(label)
plt.legend()
plt.show()
```
